Importing Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
d = pd.read_csv('nba.csv')
df = pd.DataFrame(d)

df
```

```python
df.columns
```

```python
df.describe()
```

```python
df.dtypes
```

```python
df.isnull().sum()
```

```python
df.drop(df[df.Name.isnull()].index,inplace=True)
df.isnull().sum()
```

```python
df.College.fillna('self',inplace=True)
df
```

```python
df.Salary.fillna(df.Salary.median(),inplace=True)
df.isnull().sum()
```

```python
df.describe()
```

<h3>Group By</h3>

```python
df[['Age','Team']].groupby('Team').describe()
```

```python
df[['Age','Salary']].groupby('Age').describe()
```

```python
plt.figure()

df[['Age','Salary']].groupby('Age').size().plot()

plt.show()
```

```python
d = pd.read_csv('iris.csv')
df = pd.DataFrame(d)
df
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
df.Species.value_counts()
```

```python
df.groupby('Species').describe()
```

```python
colors = {'Iris-setosa':'r', 'Iris-virginica':'g', 'Iris-versicolor':'b'}

fig, ax = plt.subplots()

for i in range(len(df['SepalLengthCm'])):
    ax.scatter(df['SepalLengthCm'][i], df['SepalWidthCm'][i],color=colors[df['Species'][i]])

ax.set_title('Iris Dataset')
ax.set_xlabel('Sepal Length')
ax.set_ylabel('Sepal Width')

plt.show()
```
