```python
import pandas as pd
```

```python
df = pd.read_csv('melb_data.csv')
df
```

```python
for col in df:
    print(col,":\n",df[col].describe())
    print()
```

```python
print("No of Dimensions",df.ndim)
print("Shape",df.shape) # Dimentionality
print("Size",df.size) #Number of elements in Dataset
print("Types: ",df.dtypes) #datatypes
```

```python
for col in df:
    print(col,": ",any(df[col].isna()))
```

```python
for col in df:
    if df[col].dtype!=object:
        df[col].fillna(df[col].mean(),inplace = True)
    else:
        df[col].fillna(df[col].mode()[0],inplace = True)
for col in df:
    print(col,": ",any(df[col].isna()))
```

```python

for col in df:
    if str(df[col].dtype) == 'int64' or str(df[col].dtype) == 'float64':
        maxi = max(df[col])
        mini = min(df[col])
        df[col] = df[col].apply(lambda x: (x-mini)/(maxi-mini))
df
```

```python
#converting categorical to quantitative 

print("*Converting Categorical to Quantitative*")
df = pd.get_dummies(df, columns=['Type','Method'])
print(df)

# for col in df:
#     if df[col].dtype == 'object' and len(df[col].unique())<=10 and (col not in ['Date','YearBuilt','Address']):
#         keys = df[col].unique()
#         values = range(0,len(keys))
#         d = dict(zip(keys, values))
#         print(d)
#         categories = [d[x] for x in df[col]]
#         df[col] = categories
# df
```
