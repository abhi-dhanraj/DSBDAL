## Data Visualization II <br>
1. Use the inbuilt dataset 'titanic' as used in the above problem. Plot a box plot for distribution of age with respect to each gender along with the information about whether they survived or not. (Column names: 'sex' and 'age')
2. Write observations on the inference from the above statistics

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('titanic')
if ~df.empty:
    print("loaded!")
```

```python
df.head(5)
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df.age.fillna(df.age.median(),inplace=True)
df.deck.fillna('C',inplace=True)
df.embarked.fillna('Q',inplace=True)
df.embark_town.fillna('Queentown',inplace=True)
```

```python
df.isnull().sum()
```

```python
plt.figure(figsize=(10,10))
sns.boxplot(data=df,x='sex',y='age',hue='survived',palette='Set3')
plt.show()
```

```python
plt.figure(figsize=(10,10))
sns.barplot(data=df,x='sex',y='age',hue='survived')
plt.show()
```

```python
plt.figure(figsize=(6,6),dpi=100)
sns.swarmplot(data=df,x='sex',y='age',hue='survived',palette='Set1')
plt.show()
```

```python
plt.figure(figsize=(6,6),dpi=100)
sns.violinplot(data=df,x='sex',y='age',hue='survived',palette='Set1')
plt.show()
```

```python
plt.figure(figsize=(6,6),dpi=100)
sns.stripplot(data=df,x='sex',y='age',hue='survived',palette='Set1')
plt.show()
```

```python
plt.figure(figsize=(6,6),dpi=100)

ax=sns.FacetGrid(data=df,row='sex',col='pclass')
ax.map(sns.histplot,'survived')

plt.show()
```
