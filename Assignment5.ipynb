```python
import pandas as pd;
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
```

```python
df=pd.read_csv('Social_Network_Ads.csv')
df
```

```python
df = df.drop('User ID',axis = 1)
df
```

```python
mapper={'Male':0, 'Female':1}
df['Gender'] = df['Gender'].apply(lambda x:mapper[x])
df
```

```python
#normalization
scaler = MinMaxScaler()
scaler.fit(df)
df = pd.DataFrame(scaler.transform(df), columns = list(df.columns))
df
```

```python
X = df[['Gender','Age','EstimatedSalary']]
y = df['Purchased']
x_train,x_test,y_train,y_test = train_test_split(X,y)
```

```python
clf = LogisticRegression()
clf.fit(x_test,y_test)
predictions = clf.predict(x_test)
report = pd.DataFrame(classification_report(y_test, predictions,output_dict = True))
report
```

```python
tp,fp,fn,tn = confusion_matrix(y_test,predictions).ravel()
print("The number of True Negatives are:{}".format(tn))
print("The number of False Positives are:{}".format(fp))
print("The number of False Negatives are:{}".format(fn))
print("The number of True Positives are:{}".format(tp))
```

```python
accuracy = (tn+tp)/(tn+tp+fn+fp)
accuracy
```

```python
error = (fn+fp)/(tn+tp+fn+fp)
error
```

```python
precision = tp/(tp+fp)
precision
```

```python
recall = tp/(tp+fn)
recall 
```
