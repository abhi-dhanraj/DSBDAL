```python
import pandas as pd #Importing libraries 
```

```python
d = pd.read_csv("melb_data.csv")  #Reading Csv file
df = pd.DataFrame(d)
```

```python
for column in df:
    print(column,":\n",df[column].describe(),end = "\n\n")
```

```python
print("No of Dimensions",df.ndim)
print("Shape",df.shape) # Dimentionality
print("Size",df.size) #Number of elements in Dataset
```

```python
for column in df:
    print(column," : ",any(df[column].isna())) #detects any missing value 
```

```python
print("*Types of variables in dataframe*")

print(df.dtypes)
```

```python
print("*Fill Missing values with 0*")
for col in df:
    if df[col].dtype!=object:
        df[col].fillna(df[col].mean(),inplace = True)
    else:
        df[col].fillna(df[col].mode()[0],inplace = True)

print("\n*Checking if there is any missing value after above operation*")
for column in df:
    print(column," : ",any(df[column].isna()))
    
```

```python
#Data normalization and Data Formatting
from sklearn.preprocessing import MinMaxScaler
import numpy as np
scaler=MinMaxScaler()
for col in df:
    if df[col].dtype!=object:
        df[col] = scaler.fit_transform(np.array(df[col]).reshape(-1, 1))

print(df)


```

```python
print("*Converting Categorical to Quantitative*")
df = pd.get_dummies(df, columns=['Type','Method'])
print(df)
```

```python
# Write to a file
print("\n* Writing to a file *")
df.to_csv('output.csv')
print("*Saved*")
```
